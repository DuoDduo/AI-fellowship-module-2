## **Exploratory Data Analysis** ##

**Importing the libraries for EDA**

```python
import pandas as pd
import os
import openpyxl
import matplotlib
import seaborn
```

```python

# Load the dataset
sales = pd.read_csv("sales_data.xls")

# Display the data
sales

# Check the data types of each column
sales.dtypes


```

```python
# Checking for missing values
sales.isna().sum()
sales.isnull().sum()
```

```python
# checking the shape of the dataframe
sales.shape
```

```python
# checking for duplicate values
sales.duplicated().sum()
sales
```

```python
# Resetting the index- method 1
sales.index = range(1, len(sales) + 1)
sales
# Resetting the index- method 2
sales_data = pd.DataFrame(sales, columns=sales.columns, index=range(1, len(sales) + 1))
sales_data
```

```python
#  this code will print the first 5 rows of customer_id, device_id and item_id columns
# [0, 2 , 4, 6]
sales_data.loc[[1, 2 , 4, 6]]
```

```python
# this code will print the first 5 rows of customer_id, device_id and item_id columns
sales_data.loc[0:5, ["customer_id", "device_id", "item_id"]]
```

```python
# Indexing columns- this will print the first 5 rows and first 3 columns
sales_data.iloc[[ 0, 2, 3, 4], [0, 1, 2]]
```

```python
# this prints all the rows and columns
sales.iloc[::]
# this prints every second row
sales.iloc[::2]
```

```python
#setting the index column of the dataset to location
sales.set_index("location", inplace = True)
sales.head()
```

```python
# this prints out the index of the dataframe which is set as the location column
sales.index 
```

```python
# this prints the unique values of the index column which is location
unique_index = sales.index.unique().sort_values()
unique_index
```

```python
# this checks the top 10 countries with highest sales
sales.index.value_counts().head(10)
```

```python
# this checks the top 10 countries with highest sales
sales.index.value_counts().head(10)

# this checks the bottom 10 countries with highest sales
sales.index.value_counts().tail(10)
```

## `Data Visualization of top and bottom 10 countries with the most sales using pandas`

```python
sales.index.value_counts().head(10).plot(kind = "bar", color = "green", title = "Top 10 countries with highest sales")
```

```python
sales.index.value_counts().tail(10).plot(kind = "bar", color = "orange", title = "Bottom 10 countries with highest sales")
```
