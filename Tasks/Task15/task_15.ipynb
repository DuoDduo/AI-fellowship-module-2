```python
# Resetting the index- method 1
sales.index = range(1, len(sales) + 1)
sales
# Resetting the index- method 2
sales_data = pd.DataFrame(sales, columns=sales.columns, index=range(1, len(sales) + 1))
sales_data
```

```python
#setting the index column of the dataset to location
sales.set_index("location", inplace = True)
sales.head()
```

```python
# this prints out the index of the dataframe which is set as the location column
sales.index 
```

```python
# this prints the unique values of the index column which is location
unique_index = sales.index.unique().sort_values()
# this checks the number of unique customer locations in the location column
unique_locations_index = sales.index.nunique()
# this prints out the number of unique locations in the location column
print(f'The number of unique customer locations is {unique_locations_index} and they are {unique_index}')
```

```python
# this checks the top 10 locations with highest sales
sales.index.value_counts().head(10)
```

```python
# this checks the top 10 locations with highest sales
sales.index.value_counts().head(10)
print(f"The top 10 locations with highest sales are {sales.index.value_counts().head(10)}")

# this checks the last 10 locations with lowest sales
sales.index.value_counts().tail(10)
print(f"The least 10 locations with lowest sales are {sales.index.value_counts().tail(10)}")
```

## `Data Visualization of top 10 locations with the most sales using pandas`
